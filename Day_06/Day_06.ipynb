Class

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    def demo(self):
        print(f"This {self.brand} is ready for demo")

car1=Car("BMW","XS")
print(car1.brand)
print(car1.model)
car1.demo()
```

```python
class User:
    def __init__(self,salary):
        self.salary=salary
    
    def addSalary(self,amount):
        self.salary+=amount
    def showSalary(self):
        print(f"Salary is: {self.salary}")

    def deductSalary(self,amount):
        self.salary-=amount

user1=User(5000)
user1.showSalary()
user1.addSalary(2000)
user1.showSalary()
user1.deductSalary(1000)
user1.showSalary()

#Directly change by calling class
user1=User(50000)
user1.showSalary()
```

```python
class animal:
    def sound(self):
        print("Animal makes sound")

class dog(animal):
    def sound(self):
        print("Dog barks")

c=dog()
c.sound()
```
