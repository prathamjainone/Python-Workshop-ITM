Dictionary

```python
#create dic
dic={
    "name":"pratham jain",
    "course":"CSE",
    "roll no":"BETN1CS23277"
}


#add
dic["semester"]="5th Sem"

#update
dic["roll no"]=5045875

rollno=dic.get("key")

print(rollno)

print(dic)

for key in dic:
    print(key,dic[key])

for key,value in dic.items():
    print(key,value)
```

```python
#make a switch calculator

a=int(input("Enter first number"))
b=int(input("Enter second number"))
op=input("Enter operator")

if(op=="+"):
    print(a+b)
elif(op=="-"):
    print(a-b)
elif(op=="*"):
    print(a*b)
elif(op=="/"):
    print(a/b)
elif(op=="%"):
    print(a%b)
elif(op=="**"):
    print(a**b)
else:
    print("Invalid operator")
```
