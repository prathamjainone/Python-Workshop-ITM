```python
temperature=int(input("What's the temperature?"))

if temperature<0:
    print("coldest")
elif temperature<10:
    print("v cold")
elif temperature<20:
    print("cold")
elif temperature<30:
    print("pleasant")
elif temperature<40:
    print("hot")
else:
    print("very hot")
```

While Loop

```python
# a=1

# while a<=10:
#     print(a)
#     a=a+1

#check palindrome code

num=int(input("tell the num"))

copy=num
ans=0
while(num>0):
    ans=ans*10+num%10
    num=num//10

if ans==copy:
    print("number is palindrome")
else:
    print("not palindrome")
```

List and Tuple

```python
nums=[1,2,2,3,4,5]
nums[3]=98
print(nums)

# for i in nums:
#     print(i)

nums.insert(2,15)
print(nums)
print(nums.count(2))
nums.sort()
nums.reverse()
print(nums)

newnums=nums.copy()
nums.clear()

print(f"{nums} and {newnums}")

t=(1,2,3,4,5,5,3)

print(t.index(4))
print(t.count(3))
```

Set

```python
s={10,2,3}
b={10,3,4}

s.pop()
print(s)
print(s.union(b))
print(s.intersection(b))
print(b.difference(s))
print(b.symmetric_difference(s)) #a^b (returns elements that are in A OR B, but NOT in both)

print(f"{s|b} and {s&b} and {b-s} and {s^b}")
```
