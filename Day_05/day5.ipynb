```python
def multiple(a,b):
    add=a+b
    sub=a-b
    mul=a*b
    div=a/b
    return add,sub,mul,div
multiple(10,20)

#ye kahi na kahi dikhega
```

```python
#recursion

def factorial(n):
    if n==1:
        return 1
    else:
        return n*factorial(n-1)
    
factorial(5)
```
